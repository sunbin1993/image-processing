```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
import math
```

```python
originImage = cv2.imread('images/lenna_original.tif')
print('shape = ',originImage.shape)
# opencv读彩色图默认是按照b g r 的格式，转成r g b 的方便处理
originImage = cv2.cvtColor(originImage, cv2.COLOR_BGR2RGB)
```

```python
def rgb2hsi(image):
    R, G, B = cv2.split(image)
    R, G, B = R/255.0, G/255.0, B/255.0
    I = (R + G + B)/3
    S = 1 - 3*np.minimum(np.minimum(R, G), B) / (R + G + B + np.finfo(np.float64).eps)
    H = np.arccos(0.5*((R-G) + (R-B)) / 
                    (np.sqrt(np.power(R-G, 2) + np.multiply(R-B, G-B)) + np.finfo(np.float64).eps)
                    )
    H[G<B] = 2*math.pi - H[G<B]
    H = H/(2*math.pi)
    H[S==0] = 0
    S[I==0] = 0
    return cv2.merge([H, S, I])    
```

```python
def hsi2rgb(hsi):
    H, S, I = cv2.split(hsi)
    H *= 2*math.pi
    B, G, R = np.zeros(H.shape), np.zeros(H.shape), np.zeros(H.shape)
    h1 = np.logical_and(H < (math.pi*2/3), H>=0)
    h2 = np.logical_and((math.pi*2/3) <= H, H < (math.pi*4/3))
    h3 = np.logical_and(H >= (math.pi*4/3), H<=2*math.pi)

    B[h1] = np.multiply(I[h1], 1-S[h1])
    R[h1] = np.multiply(I[h1], 1+np.multiply(S[h1], np.cos(H[h1]))/np.cos(math.pi/3-H[h1]))
    G[h1] = (3*I[h1] - (B[h1] + R[h1]))

    H[h2] = H[h2] - math.pi*2/3
    R[h2] = np.multiply(I[h2], 1-S[h2])
    G[h2] = np.multiply(I[h2], 1+np.multiply(S[h2], np.cos(H[h2]))/np.cos(math.pi/3-H[h2]))
    B[h2] = (3*I[h2] - (R[h2] + G[h2]))

    H[h3] = H[h3] - math.pi*4/3
    G[h3] = np.multiply(I[h3], 1-S[h3])
    B[h3] = np.multiply(I[h3], 1+np.multiply(S[h3], np.cos(H[h3]))/np.cos(math.pi/3-H[h3]))
    R[h3] = 3*I[h3] - (B[h3] + G[h3])

    return cv2.merge([B, G, R])
```

```python
def get_red_hsi(hsi):
    H, S, I = cv2.split(hsi)
    area = np.logical_not(np.logical_or(H>=(325/360), H<(23/360)))
    I[area] = 255
    return cv2.merge([H, S, I])
```

```python
hsiImg = rgb2hsi(originImage)
hsiImg = get_red_hsi(hsiImg)
rgbImg = hsi2rgb(hsiImg)
```

```python
plt.subplot(121)
plt.imshow(originImage)
plt.title("origin image")
plt.subplot(122)
plt.imshow(rgbImg)
plt.title("red component image")
plt.show()
```
