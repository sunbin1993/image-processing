```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv.imread('images/fingerprint.jpg')
plt.imshow(img)
plt.title('origin image') 
plt.show()
                         
```

```python
kernel = cv.getStructuringElement(cv.MORPH_RECT, (3,3))
print(kernel)
erode = cv.erode(img, kernel=kernel, iterations=1)
opening = cv.morphologyEx(img, op=cv.MORPH_OPEN, kernel=kernel)
dilation = cv.dilate(opening, kernel=kernel, iterations=1)
closing = cv.morphologyEx(opening, op=cv.MORPH_CLOSE, kernel=kernel)
fig, axes = plt.subplots(nrows=2, ncols=2,subplot_kw={'xticks': [], 'yticks': []},figsize=(16,6))

axes[0,0].imshow(erode)
axes[0,0].set_title('erode')
axes[0,1].imshow(opening)
axes[0,1].set_title('opening')
axes[1,0].imshow(dilation)
axes[1,0].set_title('dilation')
axes[1,1].imshow(closing)
axes[1,1].set_title('closing')
```
