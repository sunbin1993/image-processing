```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
origin_image = cv2.imread('images/color_image.jpg')
# opencvBGRmatplotlibRGB
origin_image2 = cv2.cvtColor(origin_image, cv2.COLOR_BGR2RGB)
print('origin_image2 shape',origin_image2.shape)
#
plt.figure(figsize=(20,10))
plt.imshow(origin_image2)
plt.title("origin image")
```

```python
# 
# srcImage 
# 
# return noiseImg 
def SaltAndPepper(srcImage,percentage):
    noiseImg = srcImage
    noiseNum = int(percentage * srcImage.shape[0] * srcImage.shape[1])
    print('noiseNum %s' % (noiseNum))
    for i in range (noiseNum):
        randX=np.random.randint(0,srcImage.shape[0]) 
        randY=np.random.randint(0,srcImage.shape[1]) 
        randInt = np.random.randint(0,2)
#         print('randInt',randInt)
        if randInt == 0:
            noiseImg[randX,randY] = 0
        else:
            noiseImg[randX,randY] = 255
    return noiseImg
```

```python
saltAndPepperImage = SaltAndPepper(origin_image2, 0.03)
plt.figure(figsize=(20,10))
# plt.subplot(2, 3, 2)
plt.imshow(saltAndPepperImage)
plt.title("saltAndPepper image")
```
